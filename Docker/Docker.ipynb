# Docker 
Docker is an open‑source platform used to build, package, and run applications in containers. It allows developers to bundle an application along with its dependencies so that it can run consistently across different environments.

Docker uses containerization technology, where each application runs inside an isolated container. These containers are lightweight, fast, and share the host operating system, making them more efficient than virtual machines.

For example, if an application works on my local system, I can package it using Docker and run the same container on a server or cloud without worrying about dependency issues.

⭐ DOCKER ARCHITECTURE
Docker architecture defines how different components like Docker client, Docker daemon, images, and containers work together to build and run applications in a containerized environment.

# Docker Workflow (VERY IMPORTANT ⭐)
When a user runs a command like docker run nginx, the Docker client sends the request to the Docker daemon. The daemon checks if the image is available locally; if not, it pulls the image from the Docker registry, then creates and starts the container.

# BEST FINAL ANSWER (FULL VERSION)
Docker architecture consists of several key components. The Docker client is used to send commands to the Docker daemon, which is the core engine responsible for managing containers, images, networks, and volumes. Docker images act as templates that contain the application and its dependencies, while containers are the running instances of these images. Docker registry, such as Docker Hub, is used to store and distribute images.
When a command like docker run is executed, the client communicates with the daemon, which checks for the image locally, pulls it from the registry if needed, and then creates and runs the container. This architecture enables lightweight, portable, and consistent application deployment.

# Docker LifeCycle

There are three important things,

docker build -> builds docker images from Dockerfile
docker run -> runs container from docker images
docker push -> push the container image to public/private regestries to share the docker images.

Docker daemon
The Docker daemon (dockerd) listens for Docker API requests and manages Docker objects such as images, containers, networks, and volumes. A daemon can also communicate with other daemons to manage Docker services.

Docker client
The Docker client (docker) is the primary way that many Docker users interact with Docker. When you use commands such as docker run, the client sends these commands to dockerd, which carries them out. The docker command uses the Docker API. The Docker client can communicate with more than one daemon.

Docker Desktop
Docker Desktop is an easy-to-install application for your Mac, Windows or Linux environment that enables you to build and share containerized applications and microservices. Docker Desktop includes the Docker daemon (dockerd), the Docker client (docker), Docker Compose, Docker Content Trust, Kubernetes, and Credential Helper. For more information, see Docker Desktop.

Docker registries
A Docker registry stores Docker images. Docker Hub is a public registry that anyone can use, and Docker is configured to look for images on Docker Hub by default. You can even run your own private registry.
When you use the docker pull or docker run commands, the required images are pulled from your configured registry. When you use the docker push command, your image is pushed to your configured registry. Docker objects
When you use Docker, you are creating and using images, containers, networks, volumes, plugins, and other objects. This section is a brief overview of some of those objects.

Dockerfile
Dockerfile is a file where you provide the steps to build your Docker Image.

Images
An image is a read-only template with instructions for creating a Docker container. Often, an image is based on another image, with some additional customization. For example, you may build an image which is based on the ubuntu image, but installs the Apache web server and your application, as well as the configuration details needed to make your application run.

You might create your own images or you might only use those created by others and published in a registry. To build your own image, you create a Dockerfile with a simple syntax for defining the steps needed to create the image and run it. Each instruction in a Dockerfile creates a layer in the image. When you change the Dockerfile and rebuild the image, only those layers which have changed are rebuilt. This is part of what makes images so lightweight, small, and fast, when compared to other virtualization technologies.

# Multi Stage docker builds
Multi-stage Docker build is a technique where we use multiple stages in a Dockerfile to separate the build process and runtime environment. In the first stage, we install dependencies and build the application, and in the final stage, we copy only the required artifacts into a lightweight image. This helps reduce image size, improve performance, and enhance security.

Why we use Multi-Stage Build?
 Problem (Without Multi-stage)
👉 Normal Dockerfile:
FROM python:3.9

COPY . .

RUN pip install -r requirements.txt

👉 Problem:
❌ Includes all build tools
❌ Large image size
❌ Slow deployment

Solution
👉 Use multi-stage to:
✅ Remove unnecessary files
✅ Keep only required dependencies
✅ Reduce image size

# Basic Example
Stage 1: Build
FROM python:3.9 AS builder

WORKDIR /app

COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt

Stage 2: Final
FROM python:3.9-slim

WORKDIR /app

COPY --from=builder /usr/local/lib/python3.9/site-packages /usr/local/lib/python3.9/site-packages

COPY . .

CMD ["python", "manage.py", "runserver", "0.0.0.0:8000"]
``

✅ FROM python:3.9 AS builder
👉 Starts first stage

✅ Stage 1 Purpose
Install dependencies ✅
Build app ✅


✅ FROM python:3.9-slim
👉 Second stage (lightweight image ✅)

✅ COPY --from=builder
👉 Copy only required files ✅
👉 Removes:
❌ build tools
❌ cache

What does --from=builder do?
👉It copies files from a previous build stage into the current stage.

Difference between single and multi-stage?
👉Single stage contains everything, while multi-stage separates build and runtime for optimization.

What is Distroless Image?
Distroless images are minimal Docker images that contain only the application and required runtime dependencies, without including OS tools like shells or package managers. They help reduce image size, improve security, and are mainly used in production environments. However, they make debugging difficult since no shell is available.

Distroless Dockerfile
FROM gcr.io/distroless/python3

COPY . .
CMD ["app.py"]

# Docker Binds Mounts & Volumns

In Docker, bind mounts and volumes are used for persistent storage. Bind mounts map a directory from the host system to the container, mainly used in development. Volumes are managed by Docker and provide better portability and security, making them ideal for production environments where persistent data storage is required.

 Why Storage Needed in Docker?

 👉 Containers are temporary (ephemeral)
Container stopped → data LOST ❌


👉 Example:
Django app → data stored  
Container restart → data gone ❌

✅ Solution:
Use persistent storage ✅

2. Types of Storage in Docker
✅ 1. Bind Mounts
✅ 2. Volumes

👉 These help:
Store data permanently ✅
Share data ✅

Bind Mounts :
Bind mount links a directory from the host machine to a container.

Volumes :
Volumes are managed by Docker and stored inside Docker’s storage area.

✅ Create volume
docker volume create myvolume

✅ Use volume
docker run -v myvolume:/app/data nginx

✅ Mount Volume
docker run -it -v myvolume:/app/data nginx
👉 Docker stores data in:
/var/lib/docker/volumes ✅

✅ Inspect volume
docker volumn inspect myvolumn

✅Delete volume
docker volumn rm myvolumn

👉 Docker stores:
/var/lib/docker/volumes 

✅ Q1: What is a bind mount?
👉A bind mount maps a directory from the host into the container.


✅ Q2: What is a Docker volume?
👉A volume is a Docker-managed storage used to persist data independently of containers.


✅ Q3: Difference between bind mount and volume?
👉Bind mounts are host-dependent, while volumes are managed by Docker and more portable.


✅ Q4: Why use volumes?
👉To persist data even when containers are removed or restarted.


✅ Q5: Where Docker stores volumes?
👉/var/lib/docker/volumes

✅ Q6. how to mount volumn ?
👉 To mount a volume in Docker, I use the -v option in the docker run command. For example, I can mount a host directory to a container using a bind mount or use a Docker-managed volume to store persistent data. This ensures that data is not lost when the container stops or is removed.

# Docker Networking
Docker networking enables communication between containers and external systems. By default, Docker uses a bridge network, but for better communication, we create custom networks where containers can communicate using their names. It also supports host and none network modes depending on the use case.

Types of Docker Networks
 1. Bridge Network (Default ✅)
✅ Definition:
Default network where containers communicate using internal IPs.
Ex : docker run -d nginx
👉 Runs in bridge network 
👉 Containers talk via IP 

2. Host Network
✅ Definition:
Container shares host network directly.
Ex : docker run --network host nginx
👉 No port mapping needed

3. None Network
✅ Definition:
Container has NO network.
Ex : docker run --network none nginx
👉 No internet 
👉 Fully isolated

4. Docker Network Commands

✅ List networks
docker network ls

✅ Inspect network
docker network inspect bridge

✅ Create network
docker network create mynetwork

✅ Remove network
docker network rm mynetwork
