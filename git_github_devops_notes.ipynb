# diff git fetch and git pull
git fetch is used to get the latest changes from the remote repository without affecting the working directory, while git pull fetches the changes and directly merges them into the current branch.

# HEAD 
HEAD in Git is a reference that points to the latest commit in the current branch. It tells Git where we are currently working.

# Hooks
Git hooks are scripts that run automatically before or after Git operations like commit and push. They are used to automate tasks such as code validation, testing, and enforcing rules.

Git hooks are stored locally and are not pushed to remote repositories.

In real projects, instead of local Git hooks, teams use tools like pre-commit framework and CI/CD pipelines to enforce rules and ensure code quality across all developers.

git commit → run pre-commit → then commit happens

Local Hook = before commit 
GitHub Actions = after push (server side)

Real Company Flow
Developer → Push → GitHub Actions → Security Check → Deploy


# Github Actions
GitHub Actions is used to automate CI/CD pipelines where code is tested, validated, and deployed automatically whenever developers push changes.
