# 📦 Linux Package Managers (DevOps Notes)

👉 A package manager is used to:

* Install software
* Update software
* Remove software
* Manage dependencies


# 🔹 1. `yum` (Yellowdog Updater Modified)

👉 Used in:

* CentOS 7
* RHEL (older versions)

👉 Example:

```bash
yum install nginx
yum update
yum remove nginx
```

👉 Features:

* Handles dependencies ✔
* Easy to use ✔

---

# 🔹 2. `dnf` (Dandified YUM)

👉 Used in:

* CentOS 8+
* RHEL 8+
* Fedora

👉 Replacement of `yum` ✔

👉 Example:

```bash
dnf install nginx
dnf update
dnf remove nginx
```

👉 Features:

* Faster than yum 🚀
* Better dependency management
* More stable

---

# 🔹 3. `apt` (for comparison)

👉 Used in:

* Ubuntu
* Debian

```bash
apt update
apt install nginx

# 🔹 4. `pacman`

👉 Used in:

* Arch Linux

```bash
pacman -S nginx
pacman -Syu
```

👉 Features:

* Very fast ⚡
* Simple syntax

---

# 🔹 5. `portage`

👉 Used in:

* Gentoo Linux

👉 Example:

```bash
emerge nginx
```

👉 Features:

* Source-based installation
* Highly customizable

---

# 🔹 6. `rpm` (Red Hat Package Manager)

👉 Used in:

* RHEL
* CentOS
* Fedora

👉 Works at **low level**

👉 Example:

```bash
rpm -ivh package.rpm
rpm -e package
```

👉 Features:

* Install `.rpm` files directly
* Does NOT resolve dependencies ❌

---

# 🔥 Difference Between Them

| Tool    | Level | Dependency Handling | OS        |
| ------- | ----- | ------------------- | --------- |
| yum     | High  | ✔                   | CentOS 7  |
| dnf     | High  | ✔ (better)          | CentOS 8+ |
| apt     | High  | ✔                   | Ubuntu    |
| pacman  | High  | ✔                   | Arch      |
| portage | High  | ✔                   | Gentoo    |
| rpm     | Low   | ❌                  | RHEL      |

# awk :

awk is a text processing tool
* Powerful tool to **process structured text**
* Works column-wise.
  
## 🔹 Basic Syntax

```bash
awk 'pattern {action}' file
```

## 🔹 Common Commands

### ✔ Print full file

```bash
awk '{print}' app.log
```

### ✔ Print column

```bash
awk '{print $1}' app.log
```

### ✔ Multiple columns

```bash
awk '{print $1,$2}' app.log
```

### ✔ Line number

```bash
awk '{print NR, $0}' app.log
```

### ✔ Filter lines

```bash
awk '/ERROR/ {print}' app.log
```

### ✔ Condition

```bash
awk '$3 > 100 {print}' file.txt
```

---

## 🔹 Field Separator

```bash
awk -F "," '{print $1}' file.csv
```

---

## 🔹 Built-in Variables

| Variable | Meaning          |
| -------- | ---------------- |
| $0       | full line        |
| $1,$2    | columns          |
| NR       | line number      |
| NF       | number of fields |

---
