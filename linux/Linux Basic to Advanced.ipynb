Check:

rsync --version


## 🔹 Step 4: Navigate to Windows Folder in WSL

Windows path:

C:\Users\Gunjan Patel\OneDrive\Documents\study_plan

WSL path:

cd "/mnt/c/Users/Gunjan Patel/OneDrive/Documents/study_plan"


## 🔹 Step 5: Copy .pem Key to WSL (Best Practice)

cp "/mnt/c/Users/Gunjan Patel/Downloads/linux_for_devops_key.pem" ~/
chmod 400 ~/linux_for_devops_key.pem

## 🔹 Step 6: Test SSH Connection

ssh -i ~/linux_for_devops_key.pem ubuntu@<EC2-PUBLIC-IP>

## 🔹 Step 7: Sync Files (Local → EC2)

rsync -avz -e "ssh -i ~/linux_for_devops_key.pem" linux_for_devops ubuntu@<EC2-PUBLIC-IP>:/home/ubuntu/


## 🔹 Step 8: Verify on EC2

Login to EC2:

ssh -i ~/linux_for_devops_key.pem ubuntu@<EC2-PUBLIC-IP>

Check files:

ls


# 🎯 Summary

* WSL = Linux inside Windows
* rsync = file transfer tool
* Use `/mnt/c/` for Windows files
* Always use `.pem` key for EC2 connection
* rsync uses SSH internally

# awk :

awk is a text processing tool
* Powerful tool to **process structured text**
* Works column-wise.
  
## 🔹 Basic Syntax

```bash
awk 'pattern {action}' file
```

## 🔹 Common Commands

### ✔ Print full file

```bash
awk '{print}' app.log
```

### ✔ Print column

```bash
awk '{print $1}' app.log
```

### ✔ Multiple columns

```bash
awk '{print $1,$2}' app.log
```

### ✔ Line number

```bash
awk '{print NR, $0}' app.log
```

### ✔ Filter lines

```bash
awk '/ERROR/ {print}' app.log
```

### ✔ Condition

```bash
awk '$3 > 100 {print}' file.txt
```

---

## 🔹 Field Separator

```bash
awk -F "," '{print $1}' file.csv
```

---

## 🔹 Built-in Variables

| Variable | Meaning          |
| -------- | ---------------- |
| $0       | full line        |
| $1,$2    | columns          |
| NR       | line number      |
| NF       | number of fields |

---

## 🔹 DevOps Use Cases

* Parse logs
* Extract metrics
* Process CSV/JSON-like data
  
👉 Mostly used for log analysis

🔹 Examples
✔ Print entire file
awk '{print}' app.log
✔ Print first column
awk '{print $1}' app.log
✔ Print multiple columns
awk '{print $1,$2}' app.log
✔ Print with line number
awk '{print NR, $0}' app.log
✔ Filter by word
awk '/error/ {print}' app.log
✔ Condition example
awk '$3 > 100 {print}' file.txt

Note : In awk data should be formatted like CSV,TSV.
